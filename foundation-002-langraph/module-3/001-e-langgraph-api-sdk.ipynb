## 后半部分：LangGraph API + LangGraph SDK

### 1️⃣ LangGraph API（服务端）

你需要先在 terminal 运行：

```bash
langgraph dev
```

它会启动本地 API server：`http://127.0.0.1:2024`

这个就是 **LangGraph API 本地服务端**。

### 2️⃣ LangGraph SDK（客户端）

````python
from langgraph_sdk import get_client

URL = "http://127.0.0.1:2024"
client = get_client(url=URL)
````

````python
# SDK 客户端代码 - 去调用本地的 LangGraph API
# await client.assistants.search()
# await client.threads.create()
# client.runs.stream(...)
````

---

## 🎯 最简单判断方式

| 写法 | 是什么 |
|------|------|
| `graph.stream(...)` / `graph.astream_events(...)` | **LangGraph library**（本地） |
| `client.threads.create()` / `client.runs.stream(...)` | **LangGraph SDK 调用 LangGraph API**（远程） |

---

## 💡 最终答案

> **LangGraph API** 是被 `langgraph dev` 启动的本地服务
>
> **LangGraph SDK** 是 notebook 里 `from langgraph_sdk import get_client` 创建出来的 client

前半部分 = library 本地调用 🖥️

后半部分 = SDK → API 远程调用 🌐
